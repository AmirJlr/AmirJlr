```python
import pandas as pd

books = pd.read_csv("/content/drive/MyDrive/recommender_systems/LLM/books_simple_categories.csv")
```

```python
from transformers import pipeline

classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", top_k = None)

classifier("I love this!")
```

```python
sentence_pred = classifier(books["description"][0].split("."))
sentence_pred
```

```python
for i in range(len(sentence_pred)):
    result = sentence_pred[i][0]['label']
    print(result)
```

```python
for i in sentence_pred:
    print(i[0])
```

```python
def get_emotions(sentence):
    results = classifier(sentence.split("."))

    labels = []
    for i in results:
        labels.append(i[0]['label'])

    return list(set(labels))
```

```python
get_emotions(books["description"][0])
```

```python
books["Emotion"] = books['description'].map(get_emotions)
```
